```python
from simulator import BNSimulator
from world_model import WorldAgent
from llm_agent import LLMParser

ASIA_STORY = "A patient in a chest clinic has shortness of breath. Possible causes are tuberculosis, lung cancer, bronchitis, etc. A recent trip to Asia raises TB risk. Smoking raises risk of lung cancer and bronchitis, and both TB and lung cancer can cause abnormal X-ray findings and dyspnea."
ASIA_DESCS = {
    "asia": "Visited Asia recently",
    "tub": "Has tuberculosis",
    "smoke": "Smoker",
    "lung": "Has lung cancer",
    "bronc": "Has bronchitis",
    "either": "TB or lung cancer",
    "xray": "X-ray positive",
    "dysp": "Shortness of breath",
}

sim = BNSimulator("/home/ubuntu/ai_data_scientist/BN_dataset/asia.bif")
parser = LLMParser(model_name="Qwen/Qwen2.5-1.5B-Instruct")

agent = WorldAgent(sim, ASIA_STORY, ASIA_DESCS, llm_parser=parser)

out = agent.handle("do(smoke=yes), give me 15 samples of lung and dysp")
print(out["parsed"])
display(out["data"].head())
print(out["response"])
```
